```python
# https://www.youtube.com/watch?v=_-7xizJkNCQ

import numpy as np
import matplotlib.pyplot as plt
```

```python
l = np.pi
x = np.linspace(-l, l, 500)
y = x
a_0 = 1 / l * np.trapezoid(y, x, dx=0.01)
y_fourier = np.zeros(len(x)) + a_0 / 2

for n in range(1, 10):
    a_n = 1 / l * np.trapezoid(y*np.cos(np.pi*n*x/l), x, dx=0.01)
    b_n = 1 / l * np.trapezoid(y*np.sin(np.pi*n*x/l), x, dx=0.01)

    fourier_term = a_n * np.cos(np.pi * n * x / l) + b_n * np.sin(np.pi * n * x / l)
    y_fourier = np.add(fourier_term, y_fourier)

error = np.sqrt(np.trapezoid(np.abs(np.add(y_fourier, -y))**2, x, dx=0.01))

fig = plt.figure(figsize=(8, 4.125))
ax = fig.subplots()
ax.set_aspect('equal')
ax.plot(x, y_fourier, color='black', label='Fourier Approximation')
ax.plot(x, y, '--',color='red', label='Periodic Function')
ax.set_title(f'Evaluation f(x) = x with fourier having: {n + 1} terms.\n Mean Squarred Error : {error}')
ax.set_xlabel('x')
ax.set_ylabel('f(x) - Using Fourier')
plt.xlim(1.2*min(x), 1.2*max(x))
plt.ylim(1.2*min(y), 1.2*max(y))
plt.grid()
plt.legend()
```
