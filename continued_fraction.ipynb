```python
import sympy as smp
import math
```

```python
a0 = 1 / (smp.sqrt(23) - math.floor(math.sqrt(23)))
a0
```

```python
a0s = smp.simplify(a0)
a0s
```

```python
asI = math.floor(smp.N(a0))
type(float(asI))
```

```python
a0sI = math.floor(smp.N(a0s))
a0sI
```

```python
a1 = 1 / (a0s - a0sI)
a1
```

```python
a1s = smp.simplify(a1)
a1s
```

```python
a1sI = math.floor(smp.N(a1s))
a1sI
```

```python
a2 = 1 / (a1s - a1sI)
a2
```

```python
a2s = smp.simplify(a2)
a2s
```

```python
a2sI = math.floor(smp.N(a2s))
a2sI
```

```python
a3 = 1 / (a2s - a2sI)
a3
```

```python
a3s = smp.simplify(a3)
a3s
```

```python
def get_sqrt_continued_fraction(num):
  result = []
  square_root = smp.sqrt(num)
  first_integer = math.floor(math.sqrt(num))
  start = math.floor(smp.N(first_integer + square_root))
  result.append(first_integer)

  while True:
    a_n = 1 / (square_root - first_integer)
    first_integer = smp.floor(smp.N(a_n))
    result.append(first_integer)
    square_root = a_n

    if first_integer == start:
      return result

print(get_sqrt_continued_fraction(23))
```

```python
print(get_sqrt_continued_fraction(11))
```
