```python
x = np.arange(-6, 6, 1)
y = np.arange(-6, 6, 1)
X, Y = np.meshgrid(x, y)

kq = 1
U = kq * X / (X**2 + Y**2)**(3 / 2)
V = kq * Y / (X**2 + Y**2)**(3 / 2)
N = np.hypot(U, V)

color = np.arctan2(U, V)

Q = plt.quiver(X, Y, U / N, V / N, N)
plt.colorbar(Q)
plt.axis([-6, 6, -6, 6])
plt.show()
```

```python
def fun(y, t):
    dydt = 32 - 0.16 * y
    return dydt

t0, tf, h = 0, 25, 0.1
t = np.arange(t0, tf + h, h)

ymin, ymax, ystep = 0, 400, 25
y = np.arange(ymin, ymax + ystep, ystep)

for y0 in y:
    line = odeint(fun, y0, t)
    plt.plot(t, line, 'b')

x = np.linspace(t0, tf, 20)
X, Y = np.meshgrid(x, y)

U = 1
V = fun(Y, None)
N = np.hypot(U, V)
U /= N
V /= N

plt.quiver(X, Y, U, V, angles='xy')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.axis([t0, tf, ymin, ymax])
plt.show()
```
