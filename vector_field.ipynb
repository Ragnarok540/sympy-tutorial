```python
# https://www.youtube.com/watch?v=tKgZb71BUEQ

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def fun(y, t):
    dydt = 32 - 0.16 * y
    return dydt

t0, tf, h = 0, 25, 0.1
t = np.arange(t0, tf + h, h)

ymin, ymax, ystep = 0, 400, 25
y = np.arange(ymin, ymax + ystep, ystep)

for y0 in y:
    line = odeint(fun, y0, t)
    plt.plot(t, line, 'b')

x = np.linspace(t0, tf, 20)
X, Y = np.meshgrid(x, y)

U = 1
V = fun(Y, None)
N = np.sqrt(U**2 + V**2)
U /= N
V /= N

plt.quiver(X, Y, U, V, angles='xy')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.axis([t0, tf, ymin, ymax])
plt.show()
```
