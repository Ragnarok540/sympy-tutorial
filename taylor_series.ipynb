```python
# https://www.youtube.com/watch?v=epXGq-3_CjM

import numpy as np
from math import factorial
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = np.zeros(len(x))

for n in range(0, 40):
    taylor_term = (-1)**n * x**(2 * n) / factorial(2 * n)
    y = np.add(y, taylor_term)

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('cos(x)')
ax.axis(xmin=min(x), xmax=max(x))
ax.axis(ymin=-2, ymax=2)
ax.grid()

plot, = ax.plot(x, y)
```
