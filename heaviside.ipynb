```python
from sympy import Heaviside
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 1000, endpoint=True)
y = [Heaviside(i, 1) for i in x]
```

```python
fig, ax = plt.subplots()

ax.step(x, y)
ax.set_xlim(-6, 6)
ax.set_ylim(-0.5, 1.5)
ax.grid()
```
