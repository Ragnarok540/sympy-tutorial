```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')

def plot_vectors(vectors):
  plt.figure()
  plt.grid()

  ax = plt.gca()

  max_x = 0
  max_y = 0

  for vector in vectors:
    ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1)

    if abs(vector[0]) > max_x:
      max_x = abs(vector[0])
    if abs(vector[1]) > max_y:
      max_y = abs(vector[1])

  ax.set_xlim([-max_x - 1, max_x + 1])
  ax.set_ylim([-max_y - 1, max_y + 1])

  ax.set_aspect('equal')

  plt.draw()
  plt.show()
```

```python
vector_1 = np.array([3, 2])
vector_2 = np.array([-2, 3])
```

```python
plot_vectors([vector_1])
```

```python
plot_vectors([vector_2])
```

```python
plot_vectors([vector_1, vector_2])
```
